```python
import pandas as pd
```

```python
df = pd.read_csv('car_price.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print(df['owner'].value_counts())
print(df['transmission'].value_counts())
print(df['seller_type'].value_counts())
print(df['fuel'].value_counts())
```

```python
df.replace({'owner':{'First Owner':0, 'Second Owner':1, 'Third Owner':2, 'Fourth & Above Owner':3, 'Test Drive Car':4}}, inplace=True)
df.replace({'transmission':{'Manual':0, 'Automatic':1}}, inplace=True)
df.replace({'seller_type':{'Individual':0, 'Dealer':1, 'Trustmark Dealer':2}}, inplace=True)
df.replace({'fuel':{'Diesel':0, 'Petrol':1, 'CNG':2, 'LPG':3, 'Electric':4}}, inplace=True)
df.head()
```

```python
X = df.drop(['name', 'selling_price'], axis=1)
Y = df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression, Lasso
linear = LinearRegression()
```

```python
linear.fit(X_train, Y_train)
```

```python
linear_train_predict = linear.predict(X_train)
```

```python
from sklearn.metrics import r2_score
```

```python
linear_r2_score = r2_score(Y_train, linear_train_predict)
linear_r2_score
```

```python
lasso = Lasso()
lasso.fit(X_train, Y_train)
```

```python
lasso_train_predict = lasso.predict(X_train)
lasso_r2_score = r2_score(Y_train, lasso_train_predict)
lasso_r2_score
```
